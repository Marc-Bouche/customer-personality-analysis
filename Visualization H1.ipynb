```python
#1
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a pandas DataFrame
file_path = r"C:\Users\HP\Desktop\Programming\customer_segmentation\sql-database\H1_education_marital_status_age_count_csv.csv"
df = pd.read_csv(file_path)

# Check the first few rows of the dataframe to understand its structure
print(df.head())

# Assuming the CSV has 'Education', 'age_group', 'Marital_Status', and 'people_count' columns,
# let's create a bar plot without error bars

plt.figure(figsize=(12, 8))
sns.barplot(x='age_group', y='people_count', hue='Education', data=df, errorbar=None)

# Customize the chart
plt.title('Education Level by Age Group')
plt.xlabel('Age Group')
plt.ylabel('People Count')
plt.xticks(rotation=45)
plt.legend(title='Education Level')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
#2

# Load the CSV file into a pandas DataFrame
file_path = r"C:\Users\HP\Desktop\Programming\customer_segmentation\sql-database\H1_education_spending_csv.csv"
df = pd.read_csv(file_path)

# Check the first few rows of the dataframe to understand its structure
print(df.head())

# Set the 'Education' column as the index
df.set_index('Education', inplace=True)

# Select the percentage columns for the plot
percentage_columns = ['percentage_wine', 'percentage_fruit', 'percentage_meat', 
                      'percentage_fish', 'percentage_sweet', 'percentage_promotion']

# Plot a stacked bar chart
df[percentage_columns].plot(kind='bar', stacked=True, figsize=(12, 8))

# Customize the chart
plt.title('Spending Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Percentage (%)')
plt.legend(title='Product Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
